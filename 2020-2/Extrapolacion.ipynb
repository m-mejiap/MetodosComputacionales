```python
import numpy as np
import matplotlib.pylab as plt
from scipy.optimize import curve_fit
```

```python
def pA1(x):
    return (0.5172*(x**2))+(6.7614*x)+344.62
def pB1(x):
    return (0.0383*(x**2))-(0.8954*x)-15.165
def pA2(x):
    return -(0.0048*(x**2))+(0.125*x)-0.2847
def pB2(x):
    return (0.0774*(x**2))-(1.5446*x)+3.6235
def pPHI(x):
    return (0.2266*(x**2))-(2.6137*x)+1.5268
def fit(x,A1,B1,A2,B2,PHI):
    return np.abs( (A1*np.exp(B1*x)) + (A2*np.exp( (B2+(1j*PHI))*x )) )**2

A1 = pA1(1.96)
B1 = pB1(1.96)
A2 = pA2(1.96)
B2 = pB2(1.96)
PHI = pPHI(1.96)

data = np.genfromtxt("D0data.dat")
x = data[:,0]
y = fit(x,A1,B1,A2,B2,PHI)

plt.figure(figsize=(10,7))
plt.errorbar(data[:,0],data[:,2],yerr=data[:,3],xerr=data[:,1],color="red",fmt='or',label="1.96TeV: Datos experimentales.")
plt.plot(x,y,marker='*',color="orange",label="1.96TeV: Fit")
plt.xlabel("|t|(GeV$^2$)",size=15)
plt.ylabel("dσ/dt(mb/GeV$^2$)",size=15)
plt.yscale('log')
plt.legend()
plt.savefig("5.png")
plt.show()
```

```python
newA1 = pA1(1.96)
newB1 = pB1(13)
newA2 = pA2(13)
newB2 = pB2(13)
newPHI = pPHI(13)

dat13TeV = np.genfromtxt("TOTEM13.dat")
newx = dat13TeV[:,0]
newy = fit(newx,newA1,newB1,newA2,newB2,newPHI)

plt.figure(figsize=(8,5))
plt.scatter(newx,dat13TeV[:,2],marker='*',color="red",label="13 TeV: Datos experimentales.")
plt.scatter(newx,newy,marker='*',color="orange",label="13 TeV: Fit")
plt.xlabel("|t|(GeV$^2$)",size=15)
plt.ylabel("dσ/dt(mb/GeV$^2$)",size=15)
plt.yscale('log')
plt.xlim(0,2)
plt.ylim(0.0000001,10**5)
plt.legend()
plt.savefig("132.png")
plt.show()
```
