```python
import numpy as np
import matplotlib.pylab as plt
```

```python
q = -1.6 * (10**(-19))
B = 0.4
m = 1.67 * (10**(-27))
w = (q*B)/m
v0 = 1.5 * (10**5)
theta = 0

def x(t):
    return (((-v0*np.sin(theta))*np.sin(w*t))+(v0*np.cos(theta)*(1-np.cos(w*t))))/w
def y(t):
    return (((v0*np.cos(theta))*np.sin(w*t))+(-v0*np.sin(theta)*(np.cos(w*t)-1)))/w

t = np.linspace(0,1,10000)
x = x(t)
y = y(t)
plt.plot(t,x)
plt.plot(t,y)
plt.figure(figsize=(5,5))
plt.plot(x,y)
plt.grid()
```

```python
sum = 0
int = 0
for i in range(len(y)):
    if(x[i]<0.005 and y[i]<(10**(-3)) and y[i]>(-10**(-3))):
        sum += x[i]
        int += 1
prom = sum/int
```
